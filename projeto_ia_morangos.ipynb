```python
import json
import os
from ultralytics import YOLO
from groq import Groq
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model = YOLO("/content/drive/MyDrive/IA_Morangos/best.pt")
```

```python
image_path = "/content/strawberryy.jpg"
```

```python
results = model.predict(image_path, conf=0.01, verbose=False)
```

```python
boxes = results[0].boxes

detections = []

for box in boxes:
    cls = int(box.cls[0])
    conf = float(box.conf[0])

    detections.append({
        "classe": model.names[cls],
        "confianca": round(conf, 4)
    })
```

```python
resultado_c2 = {
    "imagem": image_path,
    "total_deteccoes": len(detections),
    "deteccoes": detections
}

print(json.dumps(resultado_c2, indent=2, ensure_ascii=False))
```

```python
client = Groq(api_key="hahaha key chave")

payload = json.dumps(resultado_c2, indent=2, ensure_ascii=False)

system_prompt = "Você é um especialista em análise de qualidade de frutas."

user_prompt = f"""
Analise o resultado:

{payload}

Responda OBRIGATORIAMENTE seguindo este formato.

🍓 CLASSIFICAÇÃO
[escreva apenas uma linha]

📋 ANÁLISE

* frase 1
* frase 2
* frase 3

📊 CONFIANÇA
[Baixa, Média ou Alta]

✅ RECOMENDAÇÃO

* recomendação 1
* recomendação 2

REGRAS IMPORTANTES:

* Utilize quebras de linha.
* Nunca escreva tudo em uma única linha.
* Cada tópico deve começar em uma nova linha.
* A seção ANÁLISE deve possuir uma frase por linha.
* A seção RECOMENDAÇÃO deve possuir uma frase por linha.
* Não utilize parágrafos longos.
  """
```

```python
try:
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.2,
        max_tokens=300
    )

    print("\n=== RESPOSTA DA IA ===\n")
    print(response.choices[0].message.content)

except Exception as e:
    print("Erro:", e)
```
